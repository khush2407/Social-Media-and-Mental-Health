```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# 1) Load & prepare
df = pd.read_csv('smmh_final_preprocessed.csv')
df['HighRiskDepression'] = (df['18. How often do you feel depressed or down?'] >= 4).astype(int)

numeric_feats = [
    '1. What is your age?',
    '9. How often do you find yourself using Social media without a specific purpose?',
    '10. How often do you get distracted by Social media when you are busy doing something?',
    '11. Do you feel restless if you haven\'t used Social media in a while?',
    '12. On a scale of 1 to 5, how easily distracted are you?',
    '13. On a scale of 1 to 5, how much are you bothered by worries?',
    '14. Do you find it difficult to concentrate on things?',
    '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?',
    '16. Following the previous question, how do you feel about these comparisons, generally speaking?',
    '17. How often do you look to seek validation from features of social media?',
    '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?',
    '20. On a scale of 1 to 5, how often do you face issues regarding sleep?'
]


# 2) Train/test split
X = df[numeric_feats]
y = df['HighRiskDepression']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=23, stratify=y
)

# 3) Preprocessing: scale numerics, one‑hot encode categoricals
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_feats),
    ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_feats)
])

# 4) Define base pipelines
pipe_rf = Pipeline([('pre', preprocessor),
                    ('clf', RandomForestClassifier(random_state=42))])
pipe_gb = Pipeline([('pre', preprocessor),
                    ('clf', GradientBoostingClassifier(random_state=42))])
pipe_svc = Pipeline([('pre', preprocessor),
                     ('clf', SVC(probability=True, random_state=42))])

# 5) Hyperparameter grids for RandomizedSearch
param_rf = {
    'clf__n_estimators': [100, 200, 500],
    'clf__max_depth': [None, 10, 20, 30],
    'clf__min_samples_split': [2, 5, 10]
}
param_gb = {
    'clf__n_estimators': [100, 200, 500],
    'clf__learning_rate': [0.01, 0.1, 0.2],
    'clf__max_depth': [3, 5, 7]
}
param_svc = {
    'clf__C': [0.1, 1, 10],
    'clf__kernel': ['rbf', 'linear'],
    'clf__gamma': ['scale', 'auto']
}

# 6) Run RandomizedSearchCV
rs_rf = RandomizedSearchCV(pipe_rf, param_rf, n_iter=10, cv=5,
                           scoring='accuracy', n_jobs=-1, random_state=42)
rs_gb = RandomizedSearchCV(pipe_gb, param_gb, n_iter=10, cv=5,
                           scoring='accuracy', n_jobs=-1, random_state=42)
rs_svc = RandomizedSearchCV(pipe_svc, param_svc, n_iter=10, cv=5,
                            scoring='accuracy', n_jobs=-1, random_state=42)

rs_rf.fit(X_train, y_train)
rs_gb.fit(X_train, y_train)
rs_svc.fit(X_train, y_train)

# 7) Build a soft-voting ensemble of the three tuned models
voting = VotingClassifier(estimators=[
    ('rf', rs_rf.best_estimator_),
    ('gb', rs_gb.best_estimator_),
    ('svc', rs_svc.best_estimator_)
], voting='soft', n_jobs=-1)
voting.fit(X_train, y_train)

# 8) Evaluate
models = {
    'Random Forest (tuned)': rs_rf.best_estimator_,
    'Gradient Boosting (tuned)': rs_gb.best_estimator_,
    'SVM (tuned)': rs_svc.best_estimator_,
    'Voting Ensemble': voting
}

accuracies = {}
for name, model in models.items():
    y_pred = model.predict(X_test)
    accuracies[name] = accuracy_score(y_test, y_pred)

# 9) Print & plot
print("Model Accuracies:")
for name, acc in accuracies.items():
    print(f"{name:25} : {acc:.4f}")

# bar chart
acc_df = pd.DataFrame.from_dict(accuracies, orient='index', columns=['Accuracy'])
acc_df.sort_values('Accuracy').plot.barh(legend=False, figsize=(8,5))
plt.title("Model Comparison")
plt.xlabel("Accuracy")
plt.tight_layout()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# 1) Load your cleaned data
df = pd.read_csv("smmh.csv")

# 2) Create binary target: high-risk if Q18 ≥ 4
df["HighRiskDepression"] = (
    df["18. How often do you feel depressed or down?"] >= 4
).astype(int)

# 3) Specify feature lists (only the top-11 numerics + categoricals)
numeric_feats = [
    "1. What is your age?",
    "9. How often do you find yourself using Social media without a specific purpose?",
    "10. How often do you get distracted by Social media when you are busy doing something?",
    "11. Do you feel restless if you haven't used Social media in a while?",
    "12. On a scale of 1 to 5, how easily distracted are you?",
    "13. On a scale of 1 to 5, how much are you bothered by worries?",
    "14. Do you find it difficult to concentrate on things?",
    "16. Following the previous question, how do you feel about these comparisons, generally speaking?",
    "17. How often do you look to seek validation from features of social media?",
    "19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?",
    "20. On a scale of 1 to 5, how often do you face issues regarding sleep?"
]

categorical_feats = [
    "2. Gender",
    "3. Relationship Status",
    "4. Occupation Status",
    "6. Do you use social media?",
    "8. What is the average time you spend on social media every day?"
]

# 4) Build X and y
X = df[numeric_feats + categorical_feats]
y = df["HighRiskDepression"]

# 5) Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# 6) Preprocessing: scale numeric + OHE categorical
preprocessor = ColumnTransformer([
    ("num", StandardScaler(), numeric_feats),
    ("cat", OneHotEncoder(
        drop="first",
        handle_unknown="ignore",
        sparse_output=False
    ), categorical_feats)
], remainder="passthrough")

# 7) Pipeline and hyperparameter grid
pipe = Pipeline([
    ("pre", preprocessor),
    ("clf", LogisticRegression())  # placeholder
])

param_distributions = [
    {
      "clf": [LogisticRegression(max_iter=1000, random_state=42)],
      "clf__C": np.logspace(-3, 3, 7),
    },
    {
      "clf": [RandomForestClassifier(random_state=42)],
      "clf__n_estimators": [100, 200, 500],
      "clf__max_depth": [None, 10, 20]
    },
    {
      "clf": [GradientBoostingClassifier(random_state=42)],
      "clf__n_estimators": [100, 200, 500],
      "clf__learning_rate": [0.01, 0.1, 0.2],
      "clf__max_depth": [3, 5, 7]
    },
    {
      "clf": [SVC(probability=True, random_state=42)],
      "clf__C": [0.1, 1, 10],
      "clf__kernel": ["linear", "rbf"],
      "clf__gamma": ["scale", "auto"]
    },
    {
      "clf": [KNeighborsClassifier()],
      "clf__n_neighbors": [3, 5, 7],
      "clf__weights": ["uniform", "distance"]
    },
    {
      "clf": [XGBClassifier(
          random_state=42,
          n_jobs=1,
          eval_metric="logloss",
          verbosity=0
      )],
      "clf__n_estimators": [100, 200, 500],
      "clf__learning_rate": [0.01, 0.1, 0.2],
      "clf__max_depth": [3, 5, 7],
      "clf__subsample": [0.6, 0.8, 1.0],
      "clf__colsample_bytree": [0.6, 0.8, 1.0]
    }
]

# 8) Randomized search
search = RandomizedSearchCV(
    pipe,
    param_distributions,
    n_iter=50,
    cv=5,
    scoring="accuracy",
    n_jobs=1,
    random_state=42,
    verbose=2,
    error_score=np.nan
)
search.fit(X_train, y_train)

# 9) Results
print("▶ Best model params:", search.best_params_)
print("▶ Best CV accuracy :", search.best_score_)
y_pred = search.predict(X_test)
print("▶ Test set accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Extract the best fitted pipeline
best_pipeline = search.best_estimator_

# Get the classifier step
best_clf = best_pipeline.named_steps['clf']

# Print its class name and best parameters
print("▶ Best model:", best_clf.__class__.__name__)
print("▶ Best parameters:", search.best_params_)
print("▶ Best accuracy:", search.best_score_)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

# Assume X_test, y_test are defined, and best_pipeline is your loaded/fitted pipeline

# 1) Get predictions and predicted probabilities
y_pred = best_pipeline.predict(X_test)
y_proba = best_pipeline.predict_proba(X_test)[:, 1]  # probability for the “high-risk” class

# 2) Compute metrics
acc      = accuracy_score(y_test, y_pred)
prec     = precision_score(y_test, y_pred)
rec      = recall_score(y_test, y_pred)
f1       = f1_score(y_test, y_pred)
roc_auc  = roc_auc_score(y_test, y_proba)

# 3) Display them
print(f"Accuracy : {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")
print(f"F1 Score : {f1:.3f}")
print(f"ROC AUC  : {roc_auc:.3f}\n")

# 4) Confusion matrix and full classification report
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
